```python
import random     # library that I use in order to choose on random words from a list of words
 
name = input("What is your name? ")  # Here the user is asked to enter the name first
 
print("Wellcome ", name)
 
words = ['machine', 'learning', 'improve', 'yourself', 
         'python', 'programming', 'jupyter', 'notebook', 
         'turicreate', 'globalaihub', 'turkishaihub'] 

word = random.choice(words) # Function will choose one random word from this list of words
 
 
print("Guess the characters")
 
guesses = '' # every input character will be stored in guesses 
 
turns = 12 # number of times user can make wrong guesses
 
 
while turns > 0:
         
    failed = 0   # counts the number of times a user fails in a single word
    
    guess = input("guess a character:") # get the user guess    
    
    guesses += guess  # guesses will be accumulated

    if guess in word:
        for i in word: 
         
        # comparing that character with
        # the character in guesses
            if i in guesses: 
                print(i)
             
            else: 
                print("_")
             
            # for every failure 1 will be
            # incremented in failure
                failed += 1
             
 
        if failed == 0:
        # user will win the game if failure is 0
        # and 'You Win' will be given as output
            print("You Win") 
         
        # this print the correct word
            print("The word is: ", word) 
            break
     
    # if user has input the wrong alphabet then
    # it will ask user to enter another alphabet
   
     
    # check input with the character in word
    if guess not in word:
         
        turns -= 1
         
        # if the character doesn’t match the word
        # then “Wrong” will be given as output 
        print("Wrong")
        for i in word: 
         
        # comparing that character with
        # the character in guesses
            if i in guesses: 
                print(i)
             
            else: 
                print("_") 
        # this will print the number of
        # turns left for the user
        print("You have", + turns, 'more guesses')
         
         
        if turns == 0:
            print("You Loose")
```
