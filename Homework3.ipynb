```python
import random     # library that I use in order to choose on random words from a list of words
 
name = input("What is your name? ")  # Here the user is asked to enter the name first
 
print("Wellcome ", name)
 
words = ['machine', 'learning', 'improve', 'yourself','python', 'programming', 
         'jupyter', 'notebook','turicreate', 'globalaihub', 'turkishaihub'] 

word = random.choice(words) # Function will choose one random word from this list of words
 
print("Guess the characters")
 
guesses = '' # every input character will be stored in guesses 
 
turns = 12 # number of times user can make wrong guesses
 
while turns > 0:
    
    failed = 0 # counts the number of times a user fails in a single word
    
    guess = input("guess a character:") # get the user guess    
    
    guesses += guess  # guesses will be accumulated
    
    def printword():   # compare and print the output word
        
        global failed  # define variable as global to be able to use in a function
        
        for i in word:
            if i in guesses: 
                print(i)
                
            else: 
                print("_")
                failed += 1  # for every failure 1 will be incremented in failure
        return            
    if guess in word: # check the guess in word or not
        
        printword();  # print the output    
        
        if failed == 0:  # user will win the game if failure is 0 and 'You Win' will be given as output
            print("You Win") 
            print("The word is: ", word)  # print the correct word
            break  # end the game

            
    if guess not in word:
         
        turns -= 1   # reduce the number of guess
         
        print("Wrong")  
        
        printword();  # print the output   

        print("You have", + turns, 'more guesses')   # this will print the number of turns left for the user
         
        if turns == 0:
            print("You Loose")
```
